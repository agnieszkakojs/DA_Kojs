```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Ex 2

```python
# uniform distribution
distribut_1 = np.random.rand(1000)

# normal distribution mean = 0, standrd devation = 1
distribut_2 = np.random.normal(0, 1, 1000)

# normal distribution mean = 0, standrd devation = 10
distribut_3 = np.random.normal(0, 10, 1000)

# beta distribution for parameters below
a_b = [(1, 1), (2, 2), (3, 9)]
distribut_4 = np.random.beta(a_b[0][0], a_b[0][1], 1000)
distribut_5 = np.random.beta(a_b[1][0], a_b[1][1], 1000)
distribut_6 = np.random.beta(a_b[2][0], a_b[2][1], 1000)

# two-dimensional normal distributions
mean = [0, 0]
cov1 = [[1, 0], [0, 1]]
cov2 = [[2, 1], [1, 2]]
distribut_7 = np.random.multivariate_normal(mean, cov1, 5000).T
distribut_8 = np.random.multivariate_normal(mean, cov2, 5000).T
```

Ex 3

```python
distributs = [distribut_1, distribut_2, distribut_3, distribut_4, distribut_5, distribut_6]
titles = ["uniform", "normal", "normal", "beta", "beta", "beta"]
for i in range(6):
    bin = 25
    plt.hist(distributs[i], bin, density=True)
    plt.title("Histogram for "+titles[i]+" distibustion")
    plt.grid('on')
    plt.show()
```

Comparison of the histograms (normalized) with plots of analytical formulas for distributions.

```python
from scipy.stats import uniform, norm
bin = 15
plt.hist(distribut_1, bin, density=True)
plt.title("Histogram for "+titles[0]+" distibustion")
x = np.arange(-0.1, 1.1, 0.005)
y = uniform.pdf(x, 0, 1)
plt.plot(x, y, 'r', lw=3, alpha=0.6, label='uniform pdf') 
plt.legend(loc='best')
plt.grid('on')
plt.show()
```

```python
bin = 20
plt.hist(distribut_2, bin, density=True)
plt.title("Histogram for "+titles[1]+" distibustion; mean=0, sd=1")
x = np.arange(-3, 4, 0.01)
plt.plot(x, norm.pdf(x),'r-', lw=3, alpha=0.6, label='norm pdf')
plt.legend(loc='best')
plt.grid('on')
plt.show()
```

```python
data = np.arange(-30,30,0.01)
pdf = norm.pdf(data, loc = 0, scale = 10)
bin = 18
plt.hist(distribut_3, bin, density=True)
plt.title("Histogram for "+titles[2]+" distibustion; mean=0, sd=10")
plt.plot(data, pdf, 'r', label='norm pdf')
plt.legend(loc='best')
plt.grid('on')
plt.show()
```

```python
from scipy.stats import beta
data = np.arange(-0.02,1.02,0.001)
pdf = beta.pdf(data, 1, 1)
bin = 10
plt.hist(distribut_4, bin, density=True)
plt.title("Histogram for "+titles[3]+" distibustion; alpha=1, beta=1")
plt.plot(data, pdf, 'r', label='beta pdf')
plt.legend(loc='best')
plt.grid('on')
plt.show()
```

```python
from scipy.stats import beta
data = np.arange(-0.02,1.02,0.001)
pdf = beta.pdf(data, 2, 2)
bin = 15
plt.hist(distribut_5, bin, density=True)
plt.title("Histogram for "+titles[3]+" distibustion; alpha=2, beta=2")
plt.plot(data, pdf, 'r', label='beta pdf')
plt.legend(loc='best')
plt.grid('on')
plt.show()
```

```python
from scipy.stats import beta
data = np.arange(-0.02,1.02,0.001)
pdf = beta.pdf(data, 3, 9)
bin = 25
plt.hist(distribut_6, bin, density=True)
plt.title("Histogram for "+titles[3]+" distibustion; alpha=3, beta=9")
plt.plot(data, pdf, 'r', label='beta pdf')
plt.legend(loc='best')
plt.grid('on')
plt.show()
```

Ex 5

```python
alpha = np.random.normal(2, 0.1, 1000)
beta = np.random.normal(0.2, 0.2, 1000)
x = np.random.normal(0, 3, 1000)
sigma = np.abs(np.random.normal(0, 1, 1000))
my = alpha  + beta * x
y = np.random.normal(my, sigma)
print(y[0:10])
```

Ex 6

```python
plt.figure(figsize=(12, 6))
plt.scatter(x,y)
line_x = np.linspace(-6,6,1000)
for num in range(0,1000):
    plt.plot(line_x,line_x * beta[num] + alpha[num],'g-')
plt.title("y values as a function of x")
plt.legend(["y points", r"lines of y=$\alpha$+$\beta$*x"])
plt.show()
```
